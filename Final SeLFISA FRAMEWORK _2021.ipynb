###### Apply IQR Outliers Detection Based on Box plot

```python
def outliers(variable):
    # Calculating 1st, 3rd quartiles and iqr.
    q1, q3 = variable.quantile(0.25), variable.quantile(0.75)
    iqr = q3 - q1
    # Calculating lower fence and upper fence for outliers
    l_fence, u_fence = q1 - 1.5*iqr , q3 + 1.5*iqr   # Any values less than l_fence and greater than u_fence are outliers.
    # Observations that are outliers
    outliers = variable[(variable<l_fence) | (variable>u_fence)]
    print('Total Outliers of', variable.name,':', outliers.count())
    # Dropping outliers
    filtered = variable.drop(outliers.index, axis = 0)
    # Creating subplots
    fig, (ax1, ax2) = plt.subplots(2,1)
    # Gives space between two subplots
    fig.subplots_adjust(hspace = 1) 
    # Plotting variable with outliers
    variable.plot.box(vert = False, color = 'coral', grid = False, 
                      ax = ax1, title = 'Distribution with Outliers for %s' %variable.name)
    # Plot variable without outliers
    filtered.plot.box(vert = False, color = 'coral', grid = False, 
                      ax = ax2, title = 'Distribution without Outliers for %s' %variable.name)
```

### Create a Window Shifting Function

```python
#purpose : divide data into sequences for a given time_step , default time step is 1
#input : training variable data and responsive variable data
#output : data arranged in sequences
def create_dataset(X, y, time_steps=1):
    Xs, ys = [], []
    for i in range(len(X) - time_steps):
        v = X.iloc[i:(i + time_steps)].values
        Xs.append(v)        
        ys.append(y.iloc[i + time_steps])
    return np.array(Xs), np.array(ys)

def predict_sequence_full(model, data, window_size):
#Shift the window by 1 new prediction each time, re-run predictions on new window
    curr_frame = data[0]
    predicted = []
    for i in range(len(data)):
        predicted.append(model.predict(curr_frame[newaxis,:,:])[0,0])
        curr_frame = curr_frame[1:]
        curr_frame = np.insert(curr_frame, [window_size-1], predicted[-1], axis=0)
    return predicted

def predict_sequences_multiple(model, data, window_size, prediction_len):
#Predict sequence of <prediction_len> steps before shifting prediction run forward by <prediction_len> steps
    prediction_seqs = []
    for i in range(int(len(data)/prediction_len)):
        curr_frame = data[i*prediction_len]
        predicted = []
        for j in range(prediction_len):
            predicted.append(model.predict(curr_frame[newaxis,:,:])[0,0])
            curr_frame = curr_frame[1:]
            curr_frame = np.insert(curr_frame, [window_size-1], predicted[-1], axis=0)
        prediction_seqs.append(predicted)
    return prediction_seqs

def get_unscaled_pred_value(val, scaler, scaler_param_index):
    #This function carries out an inverse transform on a value given a scaler and the index of the feature in the scaler which val belongs
    scaler_nfeats = len(scaler.data_min_)
    feat = np.zeros((1,scaler_nfeats))
    feat[0,scaler_param_index]=val
    unsc = scaler.inverse_transform(feat)
    return unsc[0,scaler_param_index]
    
```

Determining Irregular partern of any dataset
We applied a dual approach in identifying a specific dataset that is characterised with extreme irregular patterns. This dual approach constided of both visualisation and mathematical illustration as below.

This work applied an interquartile range (IQR) calculation method to detect outliers on the sequential dataset using a box plot.  According to Myat and Win (2020), box plot illustrate the minimum and maximum values of any ordered dataset which is critical in drawing inference from data. The first quartile (Q1) of the data in box plot illustrate the median of the lower half of the data set, the median of the entire data set is the second quartile (Q2) and the third quartile (Q3) is the median of the upper half of the data set. The difference between the minimum and maximum provides a range value of the dataset. Finally, the difference between Q3 and Q1 provides an Inter-Quartile Range (IQR) given by the formula below:
IQR = Q3 - Q1.. (1)
Outliers are detected through defining through are new range called decision range as below. Data points outside the decision range are considered to be outlier. The range is as given below:
Lower Fence/Bound= (Q1 - 1.5 * IQR).. (2)
Upper Fence/Bound= (Q3 + 1.5 * IQR). (3)

Visualisation of Irregular Patterns
Billauer's algorithm for detecting local maximas and minmias in a signal (Xiao, 2018). The algorithm was further manipulated and customized to provide an irregular patterns peak period detection (IIPPD) analysis. This further provided an exploratory data analysis way to have a graphical imagery view of peaks to measure the degree of irregularity of our original data environment. The customized IIPPD algorithm detect discrete peaks values by searching for values which are surrounded by lower or larger values for maximas and minimas across the y-axis and corresponding x-axis. Look ahead value for determining the look ahead distance for a potential peak was set originally at 200-default figure. We later changed it to 25 in order to get maximum number of discrete peaks https://doi.org/10.1155/2018/2097932. The Billauer's function has a default lookahead value=200 and delta= 0

### Metrics
*Mean Absolute Error (MAE): MAE measures the average magnitude of the errors in a set of predictions, without considering their direction. MAE is more robust to data with outliers.


*Mean Squared Error represents the average of the squared difference between the original and predicted values in the data set. It measures the variance of the residuals.


*Root mean squared error (RMSE): RMSE is a quadratic scoring rule that also measures the average magnitude of the error. [MAE]  [RMSE]. The RMSE result will always be larger or equal to the MAE. If all of the errors have the same magnitude, then RMSE=MAE.
NB: Differences: Taking the square root of the average squared errors has some interesting implications for RMSE. Since the errors are squared before they are averaged, the RMSE gives a relatively high weight to large errors. This means the RMSE should be more useful when large errors are particularly undesirable.

*R Squared & Adjusted R Squared are used for explaining how well the independent variables in the linear regression model explains the variability in the dependent variable.

*They are all they are negatively-oriented scores, which means lower values are better. The lower value of MAE, MSE, and RMSE implies higher accuracy of a regression model. However, a higher value of R square is considered desirable.
